```python

# Extract year from founding_date
ai_companies['founding_year'] = ai_companies['founding_date'].dt.year

# Count companies founded each year
company_trend = ai_companies['founding_year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10,6))
plt.plot(company_trend.index, company_trend.values, marker='o')
plt.title("The Rise of AI Companies Over Time")
plt.xlabel("Founding Year")
plt.ylabel("Number of Companies Founded")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Count number of tools per company
company_tools = genAI_tools['company'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
company_tools.plot(kind='bar', color='skyblue')
plt.title("Top 10 Companies by Number of AI Tools")
plt.xlabel("Company")
plt.ylabel("Number of Tools")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Select key modality columns
modalities = ['mod_text', 'mod_image', 'mod_audio', 'mod_video']

# Group by company and sum modalities
company_modality = genAI_tools.groupby('company')[modalities].sum()

# Get top 10 most diverse companies (by total modality count)
company_modality['total_modalities'] = company_modality.sum(axis=1)
top_companies = company_modality.sort_values('total_modalities', ascending=False).head(10)

# Plot stacked bar chart
top_companies[modalities].plot(kind='bar', stacked=True, figsize=(12,7))
plt.title("Top 10 Most Diverse AI Companies by Modality")
plt.xlabel("Company")
plt.ylabel("Number of Tools Supporting Each Modality")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Modality")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select top models based on quality/cost efficiency
llm_comparison['efficiency_score'] = llm_comparison['benchmark_mmlu'] / llm_comparison['price_/_million_tokens']
top_models = llm_comparison.sort_values('efficiency_score', ascending=False).head(10)

# Replace open-source codes with labels
top_models['source_type'] = top_models['open-source'].map({1: 'Open-Source', 0: 'Proprietary'})

# Plot bar chart
plt.figure(figsize=(12,7))
sns.barplot(
    data=top_models,
    y='model',
    x='efficiency_score',
    hue='source_type',
    palette=['#00C9A7', '#FF6B6B']
)

# Add data labels
for index, row in enumerate(top_models.itertuples()):
    plt.text(row.efficiency_score + 0.1, index,
             f"{row.provider} ({row.quality_rating}/100)", 
             va='center', fontsize=10, color='white')

plt.title("⚡ Performance vs Cost: Which Models Deliver the Best Value?", fontsize=15, weight='bold')
plt.xlabel("Performance per Price Ratio (MMLU / Price per 1M Tokens)")
plt.ylabel("Model")
plt.legend(title="Model Type", loc="lower right")
plt.grid(axis='x', alpha=0.2)
plt.tight_layout()
plt.show()
```
