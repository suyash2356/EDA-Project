Analysis

```python
age_columns_incest = [
    'No. of Victims (Incest Rape Cases) upto 10 Years',
    'No. of Victims (Incest Rape Cases) - 10 - 14 Years',
    'No. of Victims (Incest Rape Cases) - 14 - 18 Years',
    'No. of Victims (Incest Rape Cases) - 18 - 30 Years',
    'No. of Victims (Incest Rape Cases) - 30 - 50 Years',
    'No. of Victims (Incest Rape Cases) above 50 Years'
]

age_columns_other = [
    'No. of Victims (Other Rape Cases) upto 10 Years',
    'No. of Victims (Other Rape Cases) - 10 - 14 Years',
    'No. of Victims (Other Rape Cases) - 14 - 18 Years',
    'No. of Victims (Other Rape Cases) - 18 - 30 Years',
    'No. of Victims (Other Rape Cases) - 30 - 50 Years',
    'No. of Victims (Other Rape Cases) above 50 Years'
]

# Sum age groups nationally (or per state)
incest_totals = df[age_columns_incest].sum()
other_totals = df[age_columns_other].sum()
```

```python
labels = ['0-10','10-14','14-18','18-30','30-50','50+']
x= range(len(labels))

plt.Figure(figsize=(10,6))
plt.bar(x,incest_totals, label ='Incest Rape Victims')
plt.bar(x,other_totals,bottom=incest_totals, label='other Rape Victims')

plt.xticks(x,labels)
plt.ylabel('Number of victims')
plt.xlabel('Age Group')
plt.title('Stacked Bar Chart of RapeVictims by Age Group and Type')
plt.legend()
plt.tight_layout()
plt.show()
```

From above chart we can understand, that most of the victims are of age group between 18-30, after them small girls(14-18) are the highest vitims. 

```python
state_cases = df.groupby('States/UTs/Cities')['No. of Cases Reported (Total Rape Cases)'].sum()
top_10_states = state_cases.sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top_10_states.plot(kind='barh', color='tomato')
plt.xlabel("No. of Cases Reported")
plt.ylabel("States/UTs")
plt.title("Top 10 states/uts with Hghest Total Rape Cases")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

From this chart we can see that overall cases in India are very high (nearly to 16000), in India the top 3 states which have highest rape case are Madhya Pradesh, followedd by Uttar Pradesh and Maharashtra.

```python
# Group data by state
grouped = df.groupby('States/UTs/Cities').sum()

# Get required columns
incest_cases = grouped['No. of Cases Reported (Incest Rape Cases)']
other_cases = grouped['No. of Cases Reported (Other Rape Cases)']

# Select top N states by total cases for clarity (optional)
top_states = (incest_cases + other_cases).sort_values(ascending=False).head(10).index

# Prepare data
incest = incest_cases[top_states]
other = other_cases[top_states]
x = np.arange(len(top_states))

# Plotting
plt.figure(figsize=(12,6))
bar_width = 0.4

plt.bar(x - bar_width/2, incest, width=bar_width, label='Incest Rape Cases', color='orange')
plt.bar(x + bar_width/2, other, width=bar_width, label='Other Rape Cases', color='steelblue')

plt.xticks(x, top_states, rotation=45, ha='right')
plt.ylabel('Number of Cases')
plt.xlabel('States/UTs')
plt.title('Comparison of Incest vs Other Rape Cases in Top 10 States')
plt.legend()
plt.tight_layout()
plt.show()
```

This chart show that only Madhya Pradesh have Incest rape case compared to other states, and other states have mostly other rape cases

```python
# Select relevant numeric columns for correlation
columns_to_use = [
    'No. of Victims (Incest Rape Cases) upto 10 Years',
    'No. of Victims (Incest Rape Cases) - 10 - 14 Years',
    'No. of Victims (Incest Rape Cases) - 14 - 18 Years',
    'No. of Victims (Incest Rape Cases) - 18 - 30 Years',
    'No. of Victims (Incest Rape Cases) - 30 - 50 Years',
    'No. of Victims (Incest Rape Cases) above 50 Years',
    'No. of Victims (Other Rape Cases) upto 10 Years',
    'No. of Victims (Other Rape Cases) - 10 - 14 Years',
    'No. of Victims (Other Rape Cases) - 14 - 18 Years',
    'No. of Victims (Other Rape Cases) - 18 - 30 Years',
    'No. of Victims (Other Rape Cases) - 30 - 50 Years',
    'No. of Victims (Other Rape Cases) above 50 Years',
    'No. of Cases Reported (Total Rape Cases)'
]

# Compute correlation matrix
corr_matrix = df[columns_to_use].corr()

# Plot heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Between Victim Age Groups and Total Rape Cases')
plt.tight_layout()
plt.show()
```

```python

# Use columns for Total Rape Cases by age group
age_columns_total = [
    'No. of Victims (Total Rape Cases) upto 10 Years',
    'No. of Victims (Total Rape Cases) - 10-14 Years',
    'No. of Victims (Total Rape Cases) - 14 - 18 Years',
    'No. of Victims (Total Rape Cases) - 18 - 30 Years',
    'No. of Victims (Total Rape Cases) - 30 - 50 Years',
    'No. of Victims (Total Rape Cases) above 50 Years'
]

# Sum the victims across all regions for each age group
age_sums = df[age_columns_total].sum()

# Labels for pie chart
labels = ['0-10', '10-14', '14-18', '18-30', '30-50', '50+']

# Plotting
plt.figure(figsize=(8,8))
plt.pie(age_sums, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Overall Victim Distribution by Age Group (Total Rape Cases in India)')
plt.tight_layout()
plt.show()
```

This pie chart describe that overall case reported in India, in which age group (18-30) are the most victims which are upto 50% of chart then small girls are the highest victims which are nearly 25%.
